```python
from scipy.stats import chi2_contingency
# plotting function and calculates p-value between 2 categorical columns to measure dependence, alpha = 0.5
def chisquare_test (df, col1, col2):
    ct = pd.crosstab(df[col1], df[col2])
    #plot all values on same plot
    ct.plot(kind='bar', figsize = (8,8))
    #plot for each entry in col1(country destination)
    fig, axes = plt.subplots(nrows = 1, ncols = ct.shape[0])
    fig.set_figheight(5)
    fig.set_figwidth(60)
    fig.tight_layout()
    
    for i in range(0 , ct.shape[0]):
        subset = ct.iloc[i,:]
        subset.plot(kind='bar', ax=axes[i])
        axes[i].set_title(subset.name)
    stat,p,dof,expected = chi2_contingency(ct)
    print(p,' ')
    if ( p <= 0.001):
        print(col1,' and ',col2, ' are dependent.')
    else:
        print(col1,' and ',col2, ' are independent.')
    return ct
```

```python
chisquare_test (df, 'country_destination', 'gender')
```
